### Univariate, Bivariate, and Multivariate Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, ax=plt.subplots(nrows=3, ncols=3, figsize=(20,20))
plt.style.use('ggplot')

ax[0][0].hist(data['Kematian_anak'],bins=20, edgecolor='k')
ax[0][0].set_xlabel('Jumlah kematian (per 1000 kelahiran)')
ax[0][0].set_ylabel('Frekuensi (Negara)')
ax[0][0].set_title('Distribusi Data Kematian Anak')

ax[0][1].hist(data['Ekspor'],bins=20,  edgecolor='k')
ax[0][1].set_xlabel('Ekspor barang perkapita')
ax[0][1].set_ylabel('Frekuensi (Negara)')
ax[0][1].set_title('Distribusi Ekspor')

ax[0][2].hist(data['Kesehatan'],bins=20,  edgecolor='k')
ax[0][2].set_xlabel('Total Pengeluaran Kesehatan perkapita')
ax[0][2].set_ylabel('Frekuensi (Negara)')
ax[0][2].set_title('Distribusi Pengeluaran Kesehatan')

ax[1][0].hist(data['Impor'],bins=20,  edgecolor='k')
ax[1][0].set_xlabel('Impor barang perkapita')
ax[1][0].set_ylabel('Frekuensi (Negara)')
ax[1][0].set_title('Distribusi Impor')

ax[1][1].hist(data['Pendapatan'],bins=20,  edgecolor='k')
ax[1][1].set_xlabel('Pendapatan perorang')
ax[1][1].set_ylabel('Frekuensi (Negara)')
ax[1][1].set_title('Distribusi Pendapatan perorang')

ax[1][2].hist(data['Inflasi'],bins=20,  edgecolor='k')
ax[1][2].set_xlabel('Tingkat Inflasi (pertumbuhan tahunan dari Total GDP)')
ax[1][2].set_ylabel('Frekuensi (Negara)')
ax[1][2].set_title('Distribusi Infasi')

ax[2][0].hist(data['Harapan_hidup'],bins=20,  edgecolor='k')
ax[2][0].set_xlabel('Rata-rata Jumlah Harapan Hidup')
ax[2][0].set_ylabel('Frekuensi (Negara)')
ax[2][0].set_title('Distribusi Harapan Hidup')

ax[2][1].hist(data['Jumlah_fertiliti'],bins=20,  edgecolor='k')
ax[2][1].set_xlabel('Rata-rata Jumlah Fertiliti')
ax[2][1].set_ylabel('Frekuensi (Negara)')
ax[2][1].set_title('Distribusi Fertiliti')

ax[2][2].hist(data['GDPperkapita'],bins=20,  edgecolor='k')
ax[2][2].set_xlabel('Rata-rata GDPperkapita (Total GDP dibagi dengan total populasi)')
ax[2][2].set_ylabel('Frekuensi (Negara)')
ax[2][2].set_title('Distribusi GDPperkapita')

plt.show()
```
